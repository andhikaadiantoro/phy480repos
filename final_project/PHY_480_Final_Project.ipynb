**Graphing Solutions of a Multivariable Differential Equation Using Different Numerical Techniques**



```python
import sympy as sym
x, t= sym.symbols('x, t')
F = -x**8 + sym.cos(4*(t**2))
F
```

**Problem:** 

The main problem is to find the graphical pattern solution of x vs. t. The solution will be represented in two different ways, the normal integer scale and the log-log scale.

**Numerical Techniques Used:**

- Euler's method
- Runge-Kutta 2nd-order
- Runge-Kutta 4th-order
- Ordinary Differential Equation Solutios To infinity




**Further description on the program:**

This program stores the equation above as a function. The x-axis points is defined as an array with a distinct step size. The y-axis points are defined as a list, in which its content are created by the for loops based the formulas of the distinct numerical techniques adding on from the initial condition of 1. At the end of the program, all of the integer-scaled graphs and the log-log scaled graphs are put together into one single graph to be compared and then the step sizes are then modified to see if there are any changes in the behavior of the graph.

**Constant Parameters**

- Lower Limit = 1.0           
- Higher Limit = 100.0          
- Step Size = 10000 (changed when comparing the 4 methods)
- Initial Condition = 1.0   

**Euler Method**



Regular f(x,t) Graph

```python
import math
from math import cos,sin,tan
from numpy import arange
import matplotlib.pyplot as plt

def f(x,t):
    return -x**8 + cos(4*(t**2)) 

low = 1.0           
high = 100.0          
step = 10000          
step_size = (high-low)/step       
ini_cond = 1.0           

x_range = arange(low,high,step_size)
y_range = []
for t in x_range:
    y_range.append(ini_cond)
    ini_cond += step_size*f(ini_cond,t)


plt.figure(figsize=(35,8))
plt.plot(x_range,y_range)
plt.title("Regular f(x,t) Graph")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```
